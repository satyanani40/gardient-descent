```python
import numpy as np
import matplotlib.pyplot as plt

# Step function definition
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Sigmoid function definition
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

# Create an array of values for x
x = np.linspace(-10, 10, 100)

# Compute values for step and sigmoid functions
step_values = step_function(x)
sigmoid_values = sigmoid_function(x)

# Plot both functions for comparison
plt.figure(figsize=(10, 5))

# Plot Step Function
plt.subplot(1, 2, 1)
plt.plot(x, step_values, label="Step Function", color='red')
plt.title("Step Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

# Plot Sigmoid Function
plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_values, label="Sigmoid Function", color='blue')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
